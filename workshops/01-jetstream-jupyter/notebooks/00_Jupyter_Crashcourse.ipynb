```python
# --- Variables and Types ---
# Python figures out the type automatically

name = "NAIRR Workshop"
year = 2026
pi = 3.14159
is_virtual = True

print(f"Welcome to {name} {year}!")
print(f"Pi is approximately {pi:.2f}")
print(f"Virtual event: {is_virtual}")
print(f"Types: {type(name).__name__}, {type(year).__name__}, {type(pi).__name__}, {type(is_virtual).__name__}")
```

```python
# --- Lists and Loops ---

systems = ["Jetstream2", "Delta", "Bridges-2", "Anvil", "Frontera"]

print("NAIRR compute systems:")
for i, system in enumerate(systems, 1):
    print(f"  {i}. {system}")

# List comprehension — a compact way to transform lists
upper_systems = [s.upper() for s in systems]
print(f"\nUppercase: {upper_systems}")

# Slicing
print(f"First two: {systems[:2]}")
print(f"Last two:  {systems[-2:]}")
```

```python
# --- Dictionaries ---

allocation = {
    "name": "NAIRR250048",
    "system": "Jetstream2",
    "institution": "Oakland University / WPI",
    "type": "Classroom",
    "cpus": 8,
    "ram_gb": 30,
}

print(f"Allocation: {allocation['name']}")
print(f"System: {allocation['system']} ({allocation['cpus']} CPUs, {allocation['ram_gb']} GB RAM)")

# Iterating over key-value pairs
print("\nAll details:")
for key, value in allocation.items():
    print(f"  {key}: {value}")
```

```python
# --- Functions ---

def compute_speedup(local_time, remote_time):
    """Calculate how much faster the remote system is."""
    speedup = local_time / remote_time
    return speedup

# Example: a task took 120 sec locally, 15 sec on Jetstream2
result = compute_speedup(120, 15)
print(f"Speedup: {result:.1f}x faster on Jetstream2")

# Try changing the numbers and re-running this cell
```

## Kernels and Environments

When you run a code cell, a **kernel** executes it. The kernel is a Python process tied to a specific **environment** — an isolated set of packages and a Python version.

**Why environments matter:**
- Different projects can need different package versions — environments keep them separate
- Share a `requirements.txt` and anyone can recreate your exact setup
- Delete the environment when done — your system Python stays clean

**Where to see your kernel:**
- **JupyterLab:** Top-right of the notebook — shows the kernel name (e.g., `workshop`)
- **VS Code:** Top-right "Select Kernel" button — lists all detected environments

**To switch between local and Jetstream2:** change the kernel. Same notebook, different kernel = code runs in a different environment with different resources.

Run the cell below to see which kernel and environment you are using right now:

```python
# --- Check Your Kernel & Environment ---
import sys
import os

print(f"Python executable:  {sys.executable}")
print(f"Conda environment:  {os.environ.get('CONDA_DEFAULT_ENV', '(not a conda env)')}")
print(f"Python version:     {sys.version.split()[0]}")
print()

# This should show the 'workshop' conda environment on Jetstream2.
# If you run this notebook locally, it will show your local Python instead.
# That's how you know which kernel is active.
```

```python
# --- Inline Plots ---
# Matplotlib plots render directly below the code cell.

# Example: timing comparison (the kind of chart you'll make in the exercises)
operations = ["Data load", "Feature eng.", "Model train", "Grid search"]
local_times = [12.3, 45.7, 89.2, 210.5]
remote_times = [3.1, 8.4, 18.6, 42.3]

x_pos = np.arange(len(operations))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(x_pos - width/2, local_times, width, label="Local laptop", color="#E74C3C")
ax.bar(x_pos + width/2, remote_times, width, label="Jetstream2", color="#2E86C1")

ax.set_ylabel("Time (seconds)")
ax.set_title("Local vs. Jetstream2 — Example Timing Comparison")
ax.set_xticks(x_pos)
ax.set_xticklabels(operations)
ax.legend()

plt.tight_layout()
plt.show()

# This is the kind of comparison you'll generate in the exercises.
```
